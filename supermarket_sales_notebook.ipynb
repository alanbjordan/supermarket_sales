### Load Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

### Load Data

```python
# load csv file into pandas data frame
df = pd.read_csv('supermarket_sales.csv')
```

### Basic Data Exploration

```python
# check first few rows to understand its structure
display(df.head())
print(df.info())
```

#### Summary Statistics

```python
numerical_columns = []
# Get numerical columns mean
for col in df.columns:
    try:
        print(col,'mean: ',round(df[col].mean(),2))
        numerical_columns.append(col)
    except:
        pass
```

```python
# Get numerical columns median
for col in numerical_columns:
    print(col,'median: ',round(df[col].median(),2))
```

```python
# Get standard deviation of numerical columns
for col in numerical_columns:
    print(col,'Std: ',round(df[col].std(),4))
```

```python
# get min of numerical columns
for col in numerical_columns:
    print(col,'min',df[col].min())

# get max of numerical columns
    print(col,'max',df[col].max(),'\n')
```

```python
# Loop through column names
for col in numerical_columns:
    # Get distribution of numerical columns
    plt.hist(df[col])
    plt.title(f'Distribution of {col}')

    plt.show()
```

#### Sales Analysis

```python
# Calculate total sales and average sales

# Total Sales
total_sales = df['Total'].sum()
print(f"Total sales: {round(total_sales,2)}") # 322966.75

# Average Sales
print(f"Average sales: {round(total_sales/len(df),2)}") # 322.97
```

```python
# Best selling products
best_sellers = df.groupby('Product line')['Total'].sum().reset_index()

# display best_sellers data frame
display(best_sellers)
```

```python
# Reformat date and time columns to date time with feature engineering

# Reformat Date column
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')

# Reformat time column
df['Time'] = pd.to_datetime(df["Time"])

# Create a copy of data frame since we are introducing new columns
df2 = df

# Create new column with day of the week
df2['Day_of_week'] = df['Date'].dt.day_name()

# Create new column with the hour of purchase
df2['Hour_of_day'] = df['Time'].dt.hour
```

```python
# Determine the busiest shoping days and hours

# Create busiest shopping days data frame
busiest_days = df['Day_of_week'].value_counts().reset_index()

# Display busiest_days
display(busiest_days)

# Create busiest shopping hours data frame
busiest_hours = df['Hour_of_day'].value_counts().reset_index()

# Display busiest_hours
display(busiest_hours)
```

### Customer Behavior Analysis

```python
# Analyze customer types distribution
customer_type = df['Customer type'].value_counts().reset_index()

# Display customer type analysis
display(customer_type)

# Analyze customer types payment methods
payment_method_by_customer = df.groupby('Customer type')['Payment'].value_counts().reset_index()

# Display payment method analysis
display(payment_method_by_customer)
```

```python
# Create member data frame
member_data = df[df['Customer type']=='Member']['Total']

# Create non member data frame
nonmember_data = df[df['Customer type']=='Normal']['Total']

# Create boxplot
plt.boxplot([member_data,nonmember_data])

# Set x-axis labels
plt.xticks([1,2], ['Member', 'Non Member'])

# Add title and and labels
plt.title('Customer Type: Statistical Comparision\n')
plt.xlabel('Customer Type')
plt.ylabel('Toal Spend')

plt.show()
```

### Visualization

```python
# Sort best_sellers by total
sorted_best_sellers = best_sellers.sort_values(by='Total')

# Create bar chart to display best sellers total spend
sorted_best_sellers.plot(kind='barh',x='Product line',y='Total',color='red')

# Add labels and title
plt.title('Best Selling Products\n')
plt.xlabel('Total Sales')
plt.ylabel('Product Line')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))

plt.show()
```

```python
# Sales by product line
pl_sales = df.groupby('Product line')['Total'].sum().reset_index()

# Product line by membertype
pl_by_customer = df.groupby(['Customer type','Product line'])['Total'].sum().reset_index()
pl_member = pl_by_customer[pl_by_customer['Customer type']=='Member']
pl_normal = pl_by_customer[pl_by_customer['Customer type']=='Normal']

# Assuming pl_sales, pl_member, and pl_normal are your DataFrames
# and that they are sorted in the same order for 'Product line'

product_lines = pl_sales['Product line']
y_pos = np.arange(len(product_lines))

bar_width = 0.3

fig, ax = plt.subplots()

# Create bars
ax.barh(y_pos + bar_width, pl_normal['Total'], bar_width, color='purple', label='Nonmember Sales', edgecolor='white', linewidth=3)
ax.barh(y_pos, pl_member['Total'], bar_width, color='gold', label='Member Sales', edgecolor='white', linewidth=3)
ax.barh(y_pos - bar_width, pl_sales['Total'], bar_width, color='grey', label='Total Sales', edgecolor='white', linewidth=3)



# Add some text for labels, title, and custom x-axis tick labels
ax.set_xlabel('Total')
ax.set_title('Sales by Product Line and Customer Type')
ax.set_yticks(y_pos)
ax.set_yticklabels(product_lines)
ax.legend(loc='upper center', bbox_to_anchor=(0.5,-0.15), ncol=3)

plt.show()

```

```python
# Product line by gender
pl_by_gender = df.groupby(['Gender','Product line'])['Total'].sum().reset_index()
pl_by_male = pl_by_gender[pl_by_gender['Gender']=='Male']
pl_by_female = pl_by_gender[pl_by_gender['Gender']=='Female']

product_lines = pl_sales['Product line']
y_pos = np.arange(len(product_lines))

bar_width = 0.3

fig, ax = plt.subplots()

# Create bars
ax.barh(y_pos + bar_width, pl_by_male['Total'], bar_width, color='blue', label='Male', edgecolor='white', linewidth=3)
ax.barh(y_pos, pl_by_female['Total'], bar_width, color='red', label='Female', edgecolor='white', linewidth=3)
ax.barh(y_pos - bar_width, pl_sales['Total'], bar_width, color='grey', label='Total Sales', edgecolor='white', linewidth=3)

# Add some text for labels, title, and custom x-axis tick labels
ax.set_xlabel('Total')
ax.set_title('Sales by Product Line and Customer Type')
ax.set_yticks(y_pos)
ax.set_yticklabels(product_lines)
ax.legend(loc='upper center', bbox_to_anchor=(0.5,-0.15), ncol=3)

plt.show()
```

```python
# Convert 'Day_of_week' to a categorical type with ordered categories
busiest_days['Day_of_week'] = pd.Categorical(busiest_days['Day_of_week'], categories=[
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
], ordered=True)

# Sort by 'Day_of_week'
sorted_day_of_week = busiest_days.sort_values('Day_of_week')

# Resize plot
plt.figure(figsize=(10,6))

# Create the bar plot
plt.bar(sorted_day_of_week['Day_of_week'], sorted_day_of_week['count'],color='red')

# Add titles and labels
plt.title(' Total Transactions by Day of the Week\n')
plt.xlabel('Day of the Week')
plt.ylabel('Transactions')

plt.show()

```

```python
# Resize figure
plt.figure(figsize=(10,6))

# Create bar chart showing transactions by hour of the day
plt.bar(busiest_hours['Hour_of_day'],busiest_hours['count'],color='red')

# Add title and labels
plt.title('Transactions by Hour\n')
plt.xlabel('Hour of the day')
plt.ylabel('Total Transactions')

plt.show()
```

```python
# Create data frame grouping data by Date and sum of sales
total_by_date = df.groupby('Date')['Total'].sum()

# Resize figure
plt.figure(figsize=(10,4))

# Create line chart for sales over time
total_by_date.plot(kind='line',x='Date',y='Total',color='red')

# Add title and labels
plt.title('Sales Over Time\n')
plt.xlabel('Date')
plt.ylabel('Total Sales')

plt.show()

```

```python
# Create a dictionary to map payment types to colors
payment_color_map = {
    'Ewallet': 'blue',
    'Cash': 'green',
    'Credit card': 'red'
}

# Initialize figure and axis
plt.figure(figsize=(15, 6))

# Loop through each payment type and plot the corresponding data
for payment_type, color in payment_color_map.items():
    payment_type_df = df[df['Payment'] == payment_type]
    payment_type_df.groupby('Date')['Total'].sum().plot(kind='line', color=color, label=payment_type)

# Add title and labels
plt.title('Sales Over Time by Payment Type')
plt.xlabel('Date')
plt.ylabel('Total Sales')

# Add legend
plt.legend(title='Payment Type')

# Show the plot
plt.show()
```

```python
# Get unique customer types for the labels of the pie chart
labels = df['Customer type'].unique()

# Get the count of each unique customer type for the sizes of the pie chart
sizes = df['Customer type'].value_counts().values

# Define the colors to be used for each customer type
colors = ['gold','purple']

# Set the figure size for the pie chart
plt.figure(figsize=(7,7))

# Create the pie chart
# 'sizes' specifies the sizes of the pie slices
# 'explode' offsets the first slice (for better visibility)
# 'labels' specifies the labels of the slices
# 'colors' specifies the colors of the slices
# 'autopct' specifies the format of the percentage to display on each slice
plt.pie(sizes, explode=(.05,0), labels=labels, colors=colors, autopct='%1.1f%%')

# Add a title to the pie chart
plt.title('Transactions by Membership Type\n')

# Add a legend to the pie chart, placing it at the 'best' location to avoid overlapping
plt.legend(labels, loc='best')

# Display the pie chart
plt.show()
```

```python
# Filter the DataFrame to only include rows where 'Customer type' is 'Member'
# Then, count the occurrences of each 'Payment' type and reset the index
df_member = df[df['Customer type']=='Member']['Payment'].value_counts().reset_index()

# Get unique payment types for the labels of the pie chart
labels = df_member['Payment'].unique()

# Get the count of each unique payment type for the sizes of the pie chart
sizes = df_member['count'].values

# Define the colors to be used for each payment type
colors = ['green','gold','red']

# Set the figure size for the pie chart
plt.figure(figsize=(7,7))

# Create the pie chart
# 'sizes' specifies the sizes of the pie slices
# 'explode' offsets the first slice (for better visibility)
# 'labels' specifies the labels of the slices
# 'colors' specifies the colors of the slices
# 'autopct' specifies the format of the percentage to display on each slice
plt.pie(sizes, explode=(0.2,0.01,0.01), labels=labels, colors=colors, autopct='%1.1f%%')

# Add a title to the pie chart
plt.title('Transaction Payment Type for Members\n')

# Add a legend to the pie chart, placing it at the 'lower left' to avoid overlapping
plt.legend(labels, loc='lower left')

# Display the pie chart
plt.show()
```
